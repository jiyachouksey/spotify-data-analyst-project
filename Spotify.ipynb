```python
import pandas as pd
```

```python
df = pd.read_csv('SpotifyFeatures.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
print (df.info())
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
top_songs = df.sort_values(by='popularity', ascending=False).head(10)
print(top_songs[['track_name', 'artist_name', 'popularity']])
```

```python
top_artists = df['artist_name'].value_counts().head(10)
print(top_artists)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['duration_ms']/60000, bins=30, kde=True) # convert ms to minutes
plt.xlabel('Duration (minutes)')
plt.title('Distribution of Song Duration')
plt.show()
```

```python
features = ['danceability','energy','tempo']
df[features].hist(bins=30, figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[features + ['popularity']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Features and Popularity')
plt.show()
```

```python
df.groupby('genre')['popularity'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 Genres by Average Popularity')
plt.ylabel('Average Popularity')
plt.show()
```
